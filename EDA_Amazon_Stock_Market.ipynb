# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

# Load Data

```python
df = pd.read_csv('https://raw.githubusercontent.com/ihsankurn/Exploratory-Data-Analysis-Amazon-Stock-Market/refs/heads/main/AMZN.csv')
df.head()
```

# Data Cleanig

```python
print(f'Size : {df.shape}')
print(f'Row : {df.shape[0]}')
print(f'Column : {df.shape[1]}')
```

```python
df.info()
```

```python
df = df.apply(pd.to_numeric, errors = 'coerce')
df.info()
```

```python
df.isnull().sum()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

# EDA

```python
df.describe()
```

```python
df.describe().T
```

```python
plt.figure(figsize = (10, 6))
sns.lineplot(data=df, x=df.index, y='Close', label='Close Label')
plt.title('Amazon Stock Market Close Price')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (10, 6))
sns.heatmap(df.corr(), annot = True, cmap = 'coolwarm', fmt='.2f')
plt.title('Correlation Map')
plt.show()
```
